### Dataset1
https://github.com/myTomorrows-research/public/tree/5b054a88746b7d4422732e2fd3ee6a77a8a53918/WWW2019


```python
dataset1_path="/content/drive/MyDrive/Viveka_datasets/expert_fully_aligned.txt"
```

```python
# Initialize a list to store the data
data = []

# Read the text file
with open(dataset1_path, 'r', encoding='utf-8') as f:
    for line in f:
        # Strip leading/trailing whitespace and skip empty lines
        line = line.strip()
        if line:
            # Split the line by tab characters
            parts = line.split('\t')
            # Ensure there are exactly two parts (medical and simple)
            if len(parts) == 2:
                data.append(parts)

# Create a DataFrame with columns 'medical' and 'simple'
df1 = pd.DataFrame(data, columns=['medical', 'simple'])
```

### Dataset3: Plaba
https://bionlp.nlm.nih.gov/plaba2023/#data

```python
# Load the JSON file
file_path = '/content/drive/MyDrive/Viveka_datasets/plaba.json'
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Initialize lists to store the extracted data
medical_list = []
simple_list = []
```

```python
for value in json_data.values():
    for inner_value in value.values():
        if 'abstract'  and 'adaptations' in inner_value:
          for key,abstract_value in inner_value['abstract'].items():
            medical_list.append(abstract_value)
            simple_value = None
            adaptations = inner_value['adaptations']
            if 'adaptation2' in adaptations and key in adaptations['adaptation2']:
                simple_value = adaptations['adaptation2'][key]
            elif 'adaptation3' in adaptations and key in adaptations['adaptation3']:
                simple_value = adaptations['adaptation3'][key]

            simple_list.append(simple_value)
```

```python
len(medical_list)
```

```python
len(simple_list)
```

```python
df3 = pd.DataFrame({
    'medical': medical_list,
    'simple': simple_list
})
```

```python
df3
```

## Dataset 4 : Paragraph level simplification
https://github.com/AshOlogn/Paragraph-level-Simplification-of-Medical-Texts/tree/main/data

```python
# Load the JSON file
file_path = '/content/drive/MyDrive/Viveka_datasets/paragraph_level_simplification.json'
with open(file_path, 'r') as file:
    json_data = json.load(file)

# Initialize lists to store the extracted data
medical_list = []
simple_list = []
```

```python
for data in json_data:
  medical_list.append(data['abstract'])
  simple_list.append(data['pls'])
```

```python
len(medical_list)
```

```python
len(simple_list)
```

```python
df4 = pd.DataFrame({
    'medical': medical_list,
    'simple': simple_list
})
```
